```python
'''Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?
Ans-Bokeh is a Python library for creating interactive visualizations and data applications in the web browser. Bokeh provides a server that allows for real-time updates to interactive plots. This server is called the Bokeh server.

The Bokeh server enables the creation of web applications with interactive visualizations that can be updated in real time. With the Bokeh server, a Python script can be used to define a plot, and this plot can be served to a web browser. The Bokeh server provides a WebSocket connection between the browser and the Python process that is running the server. This WebSocket connection allows for real-time updates to the plot.

To use the Bokeh server, you need to create a Python script that defines the plot and runs the server. Here's an example of how to create a simple Bokeh server that displays a line plot and updates it in real time:


'''
from bokeh.plotting import figure,curdoc
from random import randrange

p=figure(plot_width=400,plot_height=400)
r=p.line([0],[0])

def update():
    r.data_source.data['x'].append(randrange(100))
    r.data_source.data['y'].append(randrange(100))
curdoc().add_periodic_callback(update,1000)

curdoc().title="The real line plot"


```

```python

'''
Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?


Ans-To embed a Bokeh plot into a web page or dashboard using Flask or Django, you can use the Bokeh embed function to generate a script and div element that can be inserted into the HTML template of your web application. Here's an example of how to embed a Bokeh plot in a Flask application:


'''
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # create a plot
    plot = figure()
    plot.circle([1, 2], [3, 4])

    # generate the script and div elements for the plot
    script, div = components(plot)

    # render the template with the plot embedded
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)


# Here's an example of what the index.html template might look like:

'''<!DOCTYPE html>
<html>
  <head>
    <title>My Bokeh Plot</title>
    {{ script|safe }}
  </head>
  <body>
    <h1>My Bokeh Plot</h1>
    {{ div|safe }}
  </body>
</html>

'''
```
